```python
import os
import pandas as pd
import numpy as np
import os
import glob
import glob2
import pathlib
import re
import pyperclip
import email
```

```python
dataset_directory = "./emails/"
dataset_pathlib = pathlib.Path(dataset_directory)
all_items_in_dir = dataset_pathlib.rglob("*")
all_files_pathlib = [f for f in list(all_items_in_dir) if f.is_file()]
df = pd.DataFrame(columns=['id', 'date', 'from', 'to', 'cc', 'bcc'])
```

```python
for i, k in enumerate(all_files_pathlib):
    with open(k, 'r', encoding='windows-1252') as file_reader:
        msg = email.message_from_string(file_reader.read())
        # Fields of the email
        email_ID = msg["Message-ID"]
        date = msg["Date"]
        from_email = msg["From"]
        to = msg["To"]
        cc = msg["Cc"]
        bcc = msg["Bcc"]

        if not to: continue

        to = to.replace(" \n\t", "").replace(" ", "")

        date = re.search(r'\w+, (\d{1,2} \w{3} \d{4})', date).group(1)
        email_ID = re.search(r'<(.+).JavaMail.evans@thyme>', email_ID).group(1)
        if cc:
            cc = cc.replace(" \n\t", "").replace(" ", "")
        if bcc:
            bcc = bcc.replace(" \n\t", "").replace(" ", "")

        result = {
            "id": email_ID,
            "date": date,
            "from": from_email,
            "to": to,
            "cc": cc,
            "bcc": bcc
        }
        temp = pd.DataFrame(
            {
                'id': result['id'],
                "date": result['date'],
                "from": result["from"],
                "to": result["to"],
                "cc": result["cc"],
                "bcc": result["bcc"]
            },
            index=[0])
        df = pd.concat([df, temp])
        if i % 1000 == 0:
            print(i)
```

```python
df.to_csv("./dataset.csv")
```
